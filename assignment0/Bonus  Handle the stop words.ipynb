```python
from nltk.corpus import stopwords
words = stopwords.words('english')
print(words)
```

```python
f= open('asd.txt', 'rt',encoding='utf-8')
s=f.read()
a=(s.split())
```

```python
filtered_words = [word for word in a if word not in stopwords.words('english')]
```

```python
filtered_words
```

```python
wordcounts={}
for word in filtered_words:
    if word in wordcounts:
        wordcounts[word]+=1
    else:
        wordcounts[word]=1
s = [(k, wordcounts[k]) for k in sorted(wordcounts, key=wordcounts.get, reverse=True)]
u=s[0:15]
```

```python
u
```

```python
import csv
with open('bonus Handle the stop words.csv','w') as f:
    writer=csv.writer(f)
    header=['keywords','frenquence']
    writer.writerow(header)
    writer.writerows(u)
    f.close()
```
