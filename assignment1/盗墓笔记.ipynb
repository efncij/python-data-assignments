```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url = "http://seputu.com" 
user_agent = "Mozilla/5.0 (Windows NT 6.3; WOW64)" 
headers = {"User_agent": user_agent} 
req = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(req.text, "html.parser") 
rows = []
for mulu in soup.find_all(class_="mulu"):
    h2 = mulu.find("h2")
    if h2 is not None: 
        h2_title = h2.string 
        
        for a in mulu.find(class_="box").find_all("a"):
            href = a["href"]
            box_title = a["title"] 
            content = (h2_title, box_title, href)
            rows.append(content)        
```

```python
with open('盗墓笔记2.csv', 'w', newline='',encoding='utf-8') as fp: 
    f_csv = csv.writer(fp)
    headers_ = ('标题', '章节名', '链接')
    f_csv.writerow(headers_) 
    f_csv.writerows(rows)
```

```python
with open("盗墓笔记2.csv",encoding='utf-8') as f: 
    f_csv = csv.reader(f) 
    headers_ = next(f_csv)
    print(headers_) 
    for row in f_csv: 
        print(row)
```
