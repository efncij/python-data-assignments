```python
import urllib.request
import bs4
import csv
from bs4 import BeautifulSoup
import requests
```

```python
r=requests.get('http://cd.xiaozhu.com/')
r.text
```

```python
mypage = bs4.BeautifulSoup(r.text)
mypage
```

```python
myul = mypage.find('ul', attrs={'class':'pic_list clearfix'})
myul
```

```python
lis = myul.find_all('li')
```

```python
data = []
for li in lis:
    location = li.attrs['latlng']
    title = li.find('span', attrs={'class': 'result_title hiddenTxt'}).text
    price = li.find('span', attrs={'class': 'result_price'}).text
    comment=li.find('span', attrs={'class': 'commenthref'}).text.strip()
    span= li.find('span', attrs={'class': 'mianyajin_ico'})
    money=span.attrs['title']
    data.append([location, title, price,comment,money])
data
```
